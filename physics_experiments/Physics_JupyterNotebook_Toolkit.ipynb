## **How to Do Curve Fitting**

There are a number of tools to fit theoretical curves to data. One of the more flexible tools is called `curve_fit` from the scipy library of functions. To use this, we must define the fit function first, then pass it to the fitter along with the X and Y values we want to fit. 

```python
# This is how you do curve fitting

#First, define curve

# Defining a const + 1/r^2 fit function. The variable must always come first.
def one_over_r2(x, a0, a1):
    return a0 + a1 * pow(x,-2)  

#Perform Fit
myfit, mycov_matrix = curve_fit(one_over_r2, distance, intensity2, sigma=uncertainty)

#Printing the relevant results
print('Fit Parameters: ',myfit)
print()
print('Covariant Matrix')
print(mycov_matrix)

#What does the covariance matrix contain?

#Diagonal elements: variance of each parameter (a0, a1 in this case)
#Off-diagonal elements: covariance of each parameter

sigma = np.sqrt(np.diag(mycov_matrix))
print(f"a0: {myfit[0]:.3f} ± {sigma[0]:.3f}")
print(f"a1: {myfit[1]:.1f} ± {sigma[1]:.1f}")

#Plot the results

xp = np.linspace(0.5, 40, 70)

plt.title('Intensity vs. position with fit and Error Bars')
plt.plot(xp, [one_over_r2(x,myfit[0],myfit[1]) for x in xp], linestyle='-', color='k', label='Fit')
plt.grid()


plt.xlabel('R (cm)');
plt.ylabel('Intensity (V)');
plt.legend();

#plt.ylim(0,2)
#plt.xlim(0,30)

#plt.savefig('figure4.png')
plt.show()
```
